# INTRO

```python
from page46 import get_dataframe, get_x_values
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions


dataframe = get_dataframe()
X = get_x_values(dataframe)
plt.scatter(X[:50, 0], X[:50, 1], color="red", marker=".", label="Sentosa")
plt.scatter(X[50:100, 0], X[50:100, 1], color="blue", marker="o", label="Versicolor")
plt.xlabel("Len of field [cm]")
plt.ylabel("Len of flower [cm]")
plt.show()
```

# Perceptron

```python
import matplotlib.pyplot as plt
from page46 import Perceptron, get_dataframe, get_x_values, get_y_values


perceptron = Perceptron()
data_frame = get_dataframe()
X, y = get_x_values(data_frame), get_y_values(data_frame)
perceptron.fit(X, y)

plt.plot(range(1, len(perceptron.errors) + 1), perceptron.errors, marker="o")
plt.xlabel("Epochs")
plt.ylabel("Uptade counter")
plt.show()
```

```python
plot_decision_regions(X, y, clf=perceptron)
plt.title("Adaline - GD")
plt.xlabel("length of field")
plt.ylabel("length of flower")
plt.show()
```

# AdalineGD

```python
import matplotlib.pyplot as plt
import numpy as np

from page46 import get_dataframe, get_x_values, get_y_values
from page56 import AdalineGD


data_frame = get_dataframe()
X = get_x_values(data_frame)
y = get_y_values(data_frame)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ada1 = AdalineGD(n_iter=10, eta=0.01).fit(X, y)
ada2 = AdalineGD(n_iter=10, eta=0.0001).fit(X, y)

ax[0].plot(range(1, len(ada1.cost) + 1), np.log10(ada1.cost), marker="o")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Log (error sum)")
ax[0].set_title("Adaline 0.01")

ax[1].plot(range(1, len(ada2.cost) + 1), np.log10(ada2.cost), marker="o")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Log (error sum)")
ax[1].set_title("Adaline 0.0001")

plt.show()
```

```python
plot_decision_regions(X, y, clf=ada2)
plt.title("Adaline - GD")
plt.xlabel("length of field")
plt.ylabel("length of flower")
plt.show()
```
